```python
import numpy as np
import cv2, math, dlib, random
from matplotlib import pyplot as plt

# cap = cv2.VideoCapture('../video/chapli n.mp4')         #讀入影片
# cap = cv2.VideoCapture('../video/Alec_Baldwin.mp4')         #讀入影片
cap = cv2.VideoCapture('../video/F.mp4')         #讀影片 (影片的資料夾可以存在上課講義video的資料夾裡面)

FPS = cap.get(cv2.CAP_PROP_FPS)               # 每秒幾偵放映 Frame Per Second
F_Count = cap.get(cv2.CAP_PROP_FRAME_COUNT)   # frame count
w = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))        # 取得畫面尺寸
h = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))

seg_count = F_Count/14                   # seg=10 segmentation
print(f'FPS : {FPS:.2f} f/s,\tF_Count : {F_Count},\tw : {w},\th :　{h}')

fourcc = cv2.VideoWriter_fourcc(*'XVID')             # 使用 XVID 編碼
out = cv2.VideoWriter('../video/homework.mp4', fourcc, FPS, (w, h))  #另存影片 (名稱要換 不然會蓋過你原本的影片)
pos=(10,30);    font=0;    color=(0,255,255)

r=math.ceil((seg_count)/90)

while cap.isOpened():
    ret, frame = cap.read()   # frame  每一針回傳的照片
    if not ret  : break
    C_Count = cap.get(cv2.CAP_PROP_POS_FRAMES)

    ##### 1. Normal
    if C_Count < seg_count*1:
        # ----- your codes ---------
        cv2.putText(frame, f'{C_Count:.0f}/{F_Count:.0f} frames, FPS : {FPS:.0f} Normal', (10, 70), font, 1, color, 2)
        cv2.imshow('frame', frame)   
        cv2.waitKey(3)
        out.write(frame) # 把每一針改好的找片寫進影片
        # ---------------------------
        
    #### 2. Rotation
    elif  C_Count < seg_count*2:
        # ----- your codes ---------
        cv2.putText(frame, f'{C_Count:.0f}/{F_Count:.0f} frames, FPS : {FPS:.0f} Sobel', (10, 70), font, 1, color, 2)
        sobelx = cv2.Sobel(frame, -1, 1, 0, ksize=-1)  # 老師的程式碼上都有參數 可以調整 把原影片的frame套進去 設新的變數  
        cv2.imshow('frame', sobelx)  # show 改好的照片 新的變數 
        cv2.waitKey(3)
        out.write(sobelx) # 把每一針改好的照片(另設的變數)寫進影片
        # ---------------------------
            
    ##### 3. Sobel Canny Laplacian
    elif C_Count < seg_count*3:
        # ----- your codes ---------
        cv2.putText(frame, f'{C_Count:.0f}/{F_Count:.0f} frames, FPS : {FPS:.0f} DOG ', (10, 70), font, 1, color, 2)
        img_G0 = cv2.GaussianBlur(frame, (3, 3), 0.8, 0.8)     
        img_G1 = cv2.GaussianBlur(frame, (5, 5), 1.5, 1.5)   
        rf = img_G0 - img_G1
        # r = cv2.getRotationMatrix2D((w/2, h/2), 72, 0.8)
        # rf = cv2.warpAffine(frame, r, (w, h))
        cv2.imshow('frame', rf)
        cv2.waitKey(2) 
        out.write(rf) # 把每一針改好的找片寫進影片
        # ---------------------------
    ### 4. Sift
    elif C_Count < seg_count*4:
        # ----- your codes ---------
        cv2.putText(frame, f'{C_Count:.0f}/{F_Count:.0f} frames, FPS : {FPS:.0f} Rotation ', (10, 70), font, 1, color, 2)
        r1 = cv2.getRotationMatrix2D((w/2, h/2), 144, 0.6)
        rf1 = cv2.warpAffine(frame, r1, (w, h))
        cv2.imshow('frame', rf1)
        cv2.waitKey(1) 
        out.write(rf1) # 把每一針改好的找片寫進影片
        # ---------------------------
        
    elif C_Count < seg_count*5:
        # ----- your codes ---------
        cv2.putText(frame, f'{C_Count:.0f}/{F_Count:.0f} frames, FPS : {FPS:.0f} Rotation ', (10, 70), font, 1, color, 2)
        r2 = cv2.getRotationMatrix2D((w/2, h/2), 216, 0.4)
        rf2 = cv2.warpAffine(frame, r2, (w, h))
        cv2.imshow('frame', rf2)
        cv2.waitKey(1) 
        out.write(rf2) # 把每一針改好的找片寫進影片
        
    elif C_Count < seg_count*6:
        # ----- your codes ---------
        cv2.putText(frame, f'{C_Count:.0f}/{F_Count:.0f} frames, FPS : {FPS:.0f} Rotation ', (10, 70), font, 1, color, 2)
        r3 = cv2.getRotationMatrix2D((w/2, h/2), 288, 0.2)
        rf3 = cv2.warpAffine(frame, r3, (w, h))
        cv2.imshow('frame', rf3)
        cv2.waitKey(1) 
        out.write(rf3)# 把每一針改好的找片寫進影片
        
    elif C_Count < seg_count*7:
        # ----- your codes ---------
        cv2.putText(frame, f'{C_Count:.0f}/{F_Count:.0f} frames, FPS : {FPS:.0f} Rotation ', (10, 70), font, 1, color, 2)
        r4 = cv2.getRotationMatrix2D((w/2, h/2), 360, 1.5)
        rf4 = cv2.warpAffine(frame, r4, (w, h))
        cv2.imshow('frame', rf4)
        cv2.waitKey(1) 
        out.write(rf4)# 把每一針改好的找片寫進影片
        
    elif C_Count < seg_count*8:
#         # ----- your codes ---------
        cv2.putText(frame, f'{C_Count:.0f}/{F_Count:.0f} frames, FPS : {FPS:.0f}  Canny  ', (10, 70), font, 1, color, 2)
        canny = cv2.Canny(frame, 150, 200)
        cv2.imshow('frame', canny)
        cv2.waitKey(5) 
        out.write(canny)# 把每一針改好的找片寫進影片
        
    elif C_Count < seg_count*9:
        cv2.putText(frame, f'{C_Count:.0f}/{F_Count:.0f} frames, FPS : {FPS:.0f}  scharr  ', (10, 70), font, 1, color, 2)
        scharrx = cv2.Scharr(frame, cv2.CV_64F, 1, 0)    # try dx=1, dy=1
        scharry = cv2.Scharr(frame, cv2.CV_64F, 0, 1)

        scharrx = cv2.convertScaleAbs(scharrx)   # 轉回uint8 
        scharry = cv2.convertScaleAbs(scharry)  

        scharrxy =  cv2.addWeighted(scharrx, 0.5, scharry, 0.5, 0)
        cv2.imshow('frame', scharrxy)
        cv2.waitKey(2) 
        out.write(scharrxy)# 把每一針改好的找片寫進影片
        
    elif C_Count < seg_count*10:
        cv2.putText(frame, f'{C_Count:.0f}/{F_Count:.0f} frames, FPS : {FPS:.0f}  flip  ', (10, 70), font, 1, color, 2)
        flip_x = cv2.flip(frame, 0)
        cv2.imshow('frame', flip_x)
        cv2.waitKey(2) 
        out.write(flip_x)
            
   
    
    if cv2.waitKey(1) == 27:          # c==27 (ascii code) key escape
        break

       
        
cap.release()
out.release()
cv2.destroyAllWindows()
cv2.waitKey(1)== 27
```
